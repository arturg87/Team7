```python
import plotly
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
fiveYearDF.Date = pd.to_datetime(fiveYearDF.Date)
```

```python
fiveYearDF['month'] = fiveYearDF.Date.dt.month
```

```python
fiveYearDF['year'] = fiveYearDF.Date.dt.year
```

```python
fiveYearDF.head()
```

```python
# Sorts Date values for use in box plots.  Adapted from https://plotly.github.io/plotly.py-docs/generated/plotly.express.box.html
fiveYearSorted = fiveYearDF.sort_values('Date', ascending=True)
```

```python
fiveYearSorted.head()
```

```python
fiveYearDF.set_index('Date').index.astype('datetime64[ns]')
```

```python
px.box(fiveYearSorted, x = "month" , y = "Minimum Temperature" , color="year", hover_name ="Date", title='Monthly Minimum Temperature Range, 2017-2021')
```
